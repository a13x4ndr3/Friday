```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Keep negation words — critical for sentiment (don't, isn't, wasn't)
negation_words = {'no', 'not', 'nor', "don't", "doesn't", "didn't", "isn't",
                  "wasn't", "aren't", "weren't", "won't", "wouldn't", "can't",
                  "couldn't", "shouldn't", "never", "neither"}
stop_words = stop_words - negation_words

def preprocess_text(text):
    """Full NLP preprocessing pipeline."""
    # 1. Remove HTML tags (common in IMDB data)
    text = re.sub(r'<[^>]+>', ' ', text)
    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    # 3. Lowercase
    text = text.lower()
    # 4. Remove special characters, keep letters and spaces
    text = re.sub(r"[^a-z\s]", ' ', text)
    # 5. Tokenize (split by whitespace)
    tokens = text.split()
    # 6. Remove stop words (excluding negations) + Lemmatize
    tokens = [lemmatizer.lemmatize(t) for t in tokens if t not in stop_words and len(t) > 2]
    return ' '.join(tokens)

print("Applying preprocessing pipeline...")
start = time.time()
df['clean_review'] = df['review'].apply(preprocess_text)
elapsed = time.time() - start
print(f"Done in {elapsed:.1f}s")

# Show before/after
print("\n--- BEFORE ---")
print(df['review'].iloc[0][:300])
print("\n--- AFTER ---")
print(df['clean_review'].iloc[0][:300])
```

```python
# Encode labels: positive=1, negative=0
le = LabelEncoder()
df['label'] = le.fit_transform(df['sentiment'])
print(f"Label encoding: {dict(zip(le.classes_, le.transform(le.classes_)))}")

# Train/Test Split: 80/20 stratified
X = df['clean_review']
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
print(f"Training samples: {len(X_train):,}")
print(f"Test samples:     {len(X_test):,}")
print(f"Class balance (train): {y_train.value_counts().to_dict()}")
```

```python
# TF-IDF Vectorization
# max_features=30000 balances vocabulary richness vs. computational efficiency
# ngram_range=(1,2) captures bigrams like 'not good', 'very bad'
# sublinear_tf=True applies log normalization to dampen high-frequency terms
tfidf = TfidfVectorizer(
    max_features=30000,
    ngram_range=(1, 2),
    sublinear_tf=True,
    min_df=3,           # ignore terms appearing in fewer than 3 documents
    max_df=0.90         # ignore terms in more than 90% of documents (too common)
)

X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf  = tfidf.transform(X_test)

print(f"TF-IDF matrix shape (train): {X_train_tfidf.shape}")
print(f"TF-IDF matrix shape (test):  {X_test_tfidf.shape}")
print(f"Vocabulary size: {len(tfidf.vocabulary_):,} terms")
```

```python
results = {}  # Store all model metrics

def evaluate_model(name, model, X_tr, y_tr, X_te, y_te, sparse=True):
    """Train, evaluate, and store metrics for a given model."""
    print(f"\nTraining: {name}...")
    start = time.time()
    model.fit(X_tr, y_tr)
    train_time = time.time() - start

    y_pred = model.predict(X_te)

    acc  = accuracy_score(y_te, y_pred)
    prec = precision_score(y_te, y_pred, zero_division=0)
    rec  = recall_score(y_te, y_pred, zero_division=0)
    f1   = f1_score(y_te, y_pred, zero_division=0)

    results[name] = {
        'Accuracy': acc,
        'Precision': prec,
        'Recall': rec,
        'F1-Score': f1,
        'Train Time (s)': round(train_time, 2)
    }

    print(f"  Accuracy:  {acc:.4f}")
    print(f"  Precision: {prec:.4f}")
    print(f"  Recall:    {rec:.4f}")
    print(f"  F1-Score:  {f1:.4f}")
    print(f"  Train time: {train_time:.1f}s")
    return model
```

```python
lr_model = evaluate_model(
    'Baseline: Logistic Regression',
    LogisticRegression(C=1.0, max_iter=1000, solver='lbfgs', random_state=42),
    X_train_tfidf, y_train, X_test_tfidf, y_test
)

print("\n--- Full Classification Report ---")
y_pred_lr = lr_model.predict(X_test_tfidf)
print(classification_report(y_test, y_pred_lr, target_names=['negative', 'positive']))
```

```python
rf_model = evaluate_model(
    'Bagging: Random Forest',
    RandomForestClassifier(
        n_estimators=200,
        max_depth=None,
        max_features='sqrt',
        n_jobs=-1,
        random_state=42
    ),
    X_train_tfidf, y_train, X_test_tfidf, y_test
)

# Feature importance (top 20 tokens)
feature_names = tfidf.get_feature_names_out()
importances = rf_model.feature_importances_
top_idx = importances.argsort()[-20:][::-1]

plt.figure(figsize=(10, 5))
plt.barh([feature_names[i] for i in top_idx], importances[top_idx], color='#4CAF50')
plt.title('Random Forest — Top 20 Feature Importances', fontweight='bold')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.savefig('rf_importance.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# XGBoost with sparse matrix support
xgb_model = evaluate_model(
    'Boosting: XGBoost',
    XGBClassifier(
        n_estimators=200,
        max_depth=5,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        use_label_encoder=False,
        eval_metric='logloss',
        n_jobs=-1,
        random_state=42,
        tree_method='hist'  # faster on large sparse matrices
    ),
    X_train_tfidf, y_train, X_test_tfidf, y_test
)
```

```python
from sklearn.calibration import CalibratedClassifierCV

# Base estimators — diverse and complementary
base_estimators = [
    ('nb',  MultinomialNB(alpha=0.1)),
    ('svm', CalibratedClassifierCV(LinearSVC(C=1.0, max_iter=2000, random_state=42))),
    ('dt',  DecisionTreeClassifier(max_depth=15, random_state=42))
]

# Meta-model: Logistic Regression interprets base model probabilities
meta_model = LogisticRegression(C=1.0, max_iter=500, random_state=42)

stacking_model = evaluate_model(
    'Stacking: NB + SVM + DT → Logistic Regression',
    StackingClassifier(
        estimators=base_estimators,
        final_estimator=meta_model,
        cv=5,
        n_jobs=-1,
        passthrough=False
    ),
    X_train_tfidf, y_train, X_test_tfidf, y_test
)
```

```python
# Confusion Matrices for all models
models_dict = {
    'Logistic Regression': lr_model,
    'Random Forest': rf_model,
    'XGBoost': xgb_model,
    'Stacking': stacking_model
}

fig, axes = plt.subplots(1, 4, figsize=(20, 5))

for ax, (name, model) in zip(axes, models_dict.items()):
    y_pred = model.predict(X_test_tfidf)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
                xticklabels=['Negative', 'Positive'],
                yticklabels=['Negative', 'Positive'])
    ax.set_title(f'{name}\nAcc: {accuracy_score(y_test, y_pred):.3f}', fontweight='bold')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

plt.suptitle('Confusion Matrices — All Models', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('confusion_matrices.png', dpi=150, bbox_inches='tight')
plt.show()
```
